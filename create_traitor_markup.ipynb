```python
import pandas as pd
import io
import requests
```

```python
traitor_df = pd.read_csv("147_traitors.csv", index_col=False)
```

```python
traitor_df
```

```python
traitor_df = traitor_df.fillna('')
```

```python
traitor_df
```

```python
header = 'Name | State | Chamber | Donors | Twitter | Fired | Demands  \n'
separator = '------------ | ------------- | ------------- | ------------- | ------------- | ------------- | -------------\n'

f = open("traitors.md", "w")
f.write(header)
f.write(separator)


for _, traitor_row in traitor_df.iterrows():     

    donor_link = 'n/a'
    if traitor_row.open_secrets_id:
        donor_link = f'[Donor List]({traitor_row.top_donors_link})'
    
    f.write(f'{traitor_row.full_name} | {traitor_row.state} | {traitor_row.chamber}| {donor_link} | @{traitor_row.twitter} | no | no\n')

f.close()

#open and read the file after the appending:
f = open("traitors.md", "r")
print(f.read())
```
